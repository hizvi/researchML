#### Convert the old representation of labeled EDUs to newer

Read the labeled data

```python
edus = []

with open(r'LabeledEDUS.txt') as infile:
    for line in infile:
        tokens = line.split(' ')
        
        label = ''
        if("<negative>" in tokens): 
            label = 'n'
        elif("<positive>" in tokens): 
            label = 'p'
        elif("<neutral>" in tokens): 
            label = 'z'
            
        i = line.find('<')
        newline = line[:i] + label + '\n'
        edus.append(newline)
```

```python
edus
```

write the updated representation to file

```python
with open(r'LabeledEDUS.txt', 'w', newline='') as f:
    for line in edus:
        f.write(line)
```

test if conversion was successful

```python
l = []
with open(r'LabeledEDUS.txt') as f:
    for line in f:
        l.append(line)
        
print(l)
```
